## Data Analysis Question 1

The two most important customers of the firm may have been hacked. Verify if there are any fraudulent transactions in their history. For privacy reasons, you only know that their cardholder IDs are 2 and 18.

* Using hvPlot, create a line plot representing the time series of transactions over the course of the year for each cardholder separately. 

* Next, to better compare their patterns, create a single line plot that containins both card holders' trend data.  

* What difference do you observe between the consumption patterns? Does the difference suggest a fraudulent transaction? Explain your rationale in the markdown report.

```python
# loading data for card holder 2 from the database
# Write the query
query = """
        SELECT * FROM combined_table
        WHERE cardholder_id = 2
        AND amount < 2
        
        """

# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
cardholder_2_data = pd.read_sql(query, engine)
```

```python
# show the first 5 trasactions below $2 from card holder 2

cardholder_2_data
```

```python
# Plot for cardholder 2
cardholder_2_data.hvplot.line(
    x="date",
    y="amount",
    xlabel="Date",
    ylabel="Amount",
    title="Transactions by CARD HOLDER 2",
    rot = 90
)
```

```python
# loading data for card holder 18 from the database
# Write the query

query = """
        SELECT * FROM combined_table
        WHERE cardholder_id = 18
        AND amount < 2
        
        """

# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
cardholder_18_data = pd.read_sql(query, engine)
```

```python
# show the first 5 trasactions below $2 from card holder 2

cardholder_18_data.head()
```

```python
# Plot for cardholder 2
cardholder_18_data.hvplot.line(
    x="date",
    y="amount",
    xlabel="Date",
    ylabel="Amount",
    title="Transactions by CARD HOLDER 18",
    rot = 90
)
```

```python
# combine data from hardholders 2 and 18
# Getting data from just 2 and 18
query = """
        SELECT * FROM combined_table
        WHERE cardholder_id IN (2,18)
        AND amount < 2
        
        """

# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
cardholder_2_18_data = pd.read_sql(query, engine)
```

```python
# Displaying transactions less than $2 for both cardholder 2 and 18

display(cardholder_2_18_data.head())
display(cardholder_2_18_data.tail())
```

```python
# Combined plot for card holders 2 and 18
cardholder_2_18_data.hvplot(x = 'date', y = 'amount', by = 'cardholder_id', rot = 90, kind = 'step')
```

## Data Analysis Question 2

The CEO of the biggest customer of the firm suspects that someone has used her corporate credit card without authorization in the first quarter of 2018 to pay quite expensive restaurant bills. Again, for privacy reasons, you know only that the cardholder ID in question is 25.

* Using hvPlot, create a box plot, representing the expenditure data from January 2018 to June 2018 for cardholder ID 25.

* Are there any outliers for cardholder ID 25? How many outliers are there per month?

* Do you notice any anomalies? Describe your observations and conclusions in your markdown report.

```python
# loading data of daily transactions from jan to jun 2018 for card holder 25
# Write the query
query = """
         SELECT * FROM combined_table
        WHERE cardholder_id = 25
       
       """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
cardholder_25_data = pd.read_sql(query, engine)
```

```python
cardholder_25_data.head()
```

```python
query = """
         SELECT * 
         FROM combined_table 
         WHERE date >= '2018-01-01'  
         AND 
         date <= '2018-06-01'
         AND cardholder_id = 25
         AND merchant_cat_id = 1
       """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
new_df = pd.read_sql(query, engine)
```

```python
new_df.tail()
```

```python
# loop to change the numeric month to month names

new_df['date'] =pd.to_datetime(new_df['date'],format='%m').dt.strftime('%B')
display(new_df.head())

display(new_df.tail())
```

```python
# Creating the six box plots using hvPlot
# new_df.hvplot.box(y = 'date', groupby = 'transaction_id', ylabel='Transaction amount', height=400, width=400, legend=False)

new_df.hvplot.line(x='date', y='amount')
```

```python
new_df.hvplot.bar(x='date', y='amount')
```

```python
boxplot = new_df.hvplot.box(y='amount', by='date', height=400, width=400, legend=False)
boxplot
```
